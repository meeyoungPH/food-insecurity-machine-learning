```python
# Calculate percentage population of each group
food_access_raw_df[["TractLOWI_PCT", "TractKids_PCT", "TractSeniors_PCT", "TractWhite_PCT",
                "TractBlack_PCT", "TractAsian_PCT", "TractNHOPI_PCT", "TractAIAN_PCT",
                "TractOMultir_PCT", "TractHispanic_PCT"]] = food_access_raw_df[["TractLOWI", "TractKids", "TractSeniors", "TractWhite",
                                     "TractBlack", "TractAsian", "TractNHOPI", "TractAIAN", "TractOMultir",
                                     "TractHispanic"]]\
    .apply(lambda x: x/food_access_raw_df['Pop2010']*100).round(2)


# Display Dataframe
food_access_raw_df.head()
```

```python
# Calculate percentage TractHUNV and TractSNAP  group
food_access_raw_df[["TractHUNV_PCT", "TractSNAP_PCT"]] = food_access_raw_df[["TractHUNV", "TractSNAP"]]\
    .apply(lambda x: x/food_access_raw_df['OHU2010']*100).round(2)


# Display Dataframe
food_access_raw_df.head()
```

```python
# Create a subset dataframe with selected columns
first_food_access_df = food_access_raw_df[["Pop2010", "OHU2010", "PovertyRate", "MedianFamilyIncome", "TractLOWI", 
                                    "LAhalfand10", "lapophalfshare", "lalowihalfshare", "lakidshalfshare", 
                                    "laseniorshalfshare", "lawhitehalfshare", "lablackhalfshare", 
                                    "laasianhalfshare", "lanhopihalfshare", "laaianhalfshare", "laomultirhalfshare",
                                    "lahisphalfshare", "lahunvhalfshare", "lasnaphalfshare", "TractLOWI_PCT", 
                                    "TractKids_PCT", "TractSeniors_PCT", "TractWhite_PCT", "TractBlack_PCT", 
                                    "TractAsian_PCT", "TractNHOPI_PCT", "TractAIAN_PCT", "TractOMultir_PCT", 
                                    "TractHispanic_PCT", "TractHUNV_PCT", "TractSNAP_PCT"]]
first_food_access_df.tail()
```

```python
# Drop rows with NAN values
first_food_access_df = first_food_access_df.dropna()
 
# Reset the index
first_food_access_df = first_food_access_df.reset_index(drop = True)

first_food_access_df.tail()
```

### SPLIT AND NORMALIZE

```python
# Split our preprocessed data into our features and target arrays
y = first_food_access_df['LAhalfand10']
X = first_food_access_df.drop(['LAhalfand10'], axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### TRAIN

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  8
hidden_nodes_layer2 = 12

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(hidden_nodes_layer1, activation="relu", input_dim= 30))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(hidden_nodes_layer2, activation="relu"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=80)
```

### RESULTS

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

In this case, a loss of 0.3283 suggests that the model is making relatively accurate predictions, while an accuracy of 0.8462 suggests that the model is correct in its predictions for approximately 85% of the samples in the dataset. Further optimization such as adding hyperparameter tuning could potentially improve its generalizability, efficiency, or performance on new datasets or tasks.
